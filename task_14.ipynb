```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


def create_abstract_pattern(width, height):
    image = np.zeros((height, width, 3), dtype=np.uint8)

    for y in range(height):
        for x in range(width):
            r = np.sin(0.1 * x) * 127 + 128
            g = np.sin(0.1 * y) * 127 + 128
            b = np.cos(0.1 * x + 0.1 * y) * 127 + 128
            image[y, x] = [r, g, b]

    return image

# Define image dimensions
width, height = 400, 400

# Generate the abstract pattern
abstract_image = create_abstract_pattern(width, height)

# Display the image using Matplotlib
plt.imshow(abstract_image)
plt.axis('off')

# Save the image to a file (change 'abstract_pattern.png' to your desired filename and format)
plt.savefig('abstract_pattern.png')

# Display the image (optional)
plt.show()


```
